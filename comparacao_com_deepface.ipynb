```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
def verificacao(caminho_imagem1, caminho_imagem2):
  imagem1 = cv2.imread(caminho_imagem1)
  imagem2 = cv2.imread(caminho_imagem2)

  if imagem1 is None or imagem2 is None:
    print("Erro ao carregar as imagens.")
    return False

  plt.imshow(imagem1[:, :, ::-1])
  plt.title("Imagem 1")
  plt.show()
  plt.imshow(imagem2[:, :, ::-1])
  plt.title("Imagem 2")
  plt.show()

  resultado = DeepFace.verify(caminho_imagem1, caminho_imagem2)
  print(f'Resultado: {resultado}\n')

  verificacao_imagens = resultado["verified"]
  if verificacao_imagens:
    print('As imagens são da mesma pessoa')
  else:
    print('As imagens não são da mesma pessoa')
```

```python
verificacao("zeca-pagodinho.jpeg", "zeca2.png")
```

```python
verificacao("arlindo.jpeg", "zeca-pagodinho.jpeg")
```
